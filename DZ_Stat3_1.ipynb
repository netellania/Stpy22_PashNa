**Задача.** Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.

Вопросы:

Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?

```python
# Считаем,что игра выгодна игроку, когда потратив 1 рубльна игру по итогам он получает любой выигрыш.
# Т.е., если он и не остался в плючсе,то хотябы компенсировал свои расходы.
#Это события: 
#событие 1 777: 200 руб.
#событие 2 999: 100 руб.
#событие 3 555: 50 руб.
#событие 4 333: 15 руб.
#событие 5 111: 10 руб.
#событие 6 *77: 5 руб.
#событие 7 **7: 3 руб.
#событие 8 *00: 2 руб.
#событие 9 **0: 1 руб.

#Все пространство событий составляет от 0 до 999, т.е. 1000
#события с первого по пятое имеют вероятность P(A1-5)=1/1000=0.001, т.к. такие значения входят в пространстов событий по1 разу.
#событие *77 составляет вероятность P(A6)=8/1000=0.008, так как такие числа встречаются чаще, берем 8 раз,так как сюда входит значение 777 из события 1, исключаем дубли.
#событие **7 имеет вероятность P(A7)=81/1000=0.081, так как такие числа встречаются чаще,но берем 81,
#так как так как сюда входят 9 значений события 6,учитывая событие 1.
#событие *00 имеет вероятность P(A8)=9/1000=0.009, так как такие числа встречаются чаще 9 раз
# количество успехов в пространстве событий 5+8+81+9=103
#событие **0 имеет вероятность P(A9)=81/1000=0.081, так как такие числа встречаются чаще 9 раз
# количество успехов в пространстве событий 5+8+81+9+81=184
# Вероятность успеха P(A)=0.001(событие 1)+0.001(событие 2)+0.001(событие 3)+0.001(событие 4)+0.001(событие 5)+0,008(событие 6)+
# 0,081(событие 7)+ 0,009(событие 8)+0,081(событие 9)=0,184 


import numpy as np

trials = 1000    # Пространство событий
successes = 0    # Количество успехов
summ=0           # Сумма выигрыша
sl_chis=0        # Счетчик чисел
suc77=[177, 277,377, 477,577,677,877,977]
suc00=[100, 200,300, 400,500,600,700,800,900]

for i in range(trials):
    random_score = np.random.randint(1, 999)    # Случайное число 
    if random_score==777:
        successes += 1 
        summ+=200
        #print(random_score)
    if random_score==999:
        successes += 1 
        summ+=100
        #print(random_score)
    if random_score==555:
        successes += 1 
        summ+=50    
        #print(random_score)
    if random_score==333:
        successes += 1 
        summ+=15 
        #print(random_score)
    if random_score==111:
        successes += 1 
        summ+=10 
        #print(random_score)
    if random_score in suc77:
        successes += 1
        summ+=5
        #print(random_score)
    if random_score in suc00:
        successes += 1
        summ+=2
        #print(random_score)
    if random_score%10==7 and random_score//100>10 and (random_score//100)%10!=7: 
        successes += 1  
        summ+=3
        #print(random_score)
    if random_score%10==0 and random_score//100>10 and (random_score//100)%10>0: 
        successes += 1  
        summ+=1
        #print(random_score)
    
        
p_monte_carlo = successes / trials    # Вероятность из моделирования
p = 184 / 1000                        # Теоретическая вероятность
sr= summ/trials                       # Средний выигрыш
sr_plus=trials - summ



print('Количество успехов', successes)
print('Сумма выигрыша', summ)
print('Пространство событий',trials)
print('Вероятность выигрыша при моделировании',p_monte_carlo)
print('Теоретическая вероятность', p)
print('Средний выигрыш', sr)

# так как вероятность выигрыша при такоих условиях меньше,чем теоретическая вероятность, то можно говорить о том,что игроку играть не выгодно
# при максимальновозможной генерации событий в 1000 игрок потратить 1000 рублей за каждый запуск, а максимально возможный выигрыш составит 
# 200+100+50+15+10+40+243+18+81=757р
# в среднем за игру игрок приобретает меньше рубля

```
